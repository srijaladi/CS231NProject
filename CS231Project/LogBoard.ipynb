```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
train_loss_np = np.loadtxt("DataFolder/Logging/train_losses.txt")
val_loss_np   = np.loadtxt("DataFolder/Logging/val_losses.txt")
val_acc_np    = np.loadtxt("DataFolder/Logging/val_accuracies.txt")

x_axis = np.arange(len(train_loss_np))
plt.clf()
plt.plot(x_axis, train_loss_np)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Training Loss over iterations")
plt.show()
plt.clf()

plt.clf()
plt.plot(x_axis, val_loss_np)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Validation Loss over iterations")
plt.show()
plt.clf()

plt.clf()
plt.plot(x_axis, val_acc_np)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Validation Modified Accuracy over iterations")
plt.show()
plt.clf()
```
